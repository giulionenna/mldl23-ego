```python
import pandas as pd
import hickle
import numpy as np
import pickle
```

```python
def writePickle(filename,a):
    with open(filename, 'wb') as handle:
        pickle.dump(a, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
def addElementToDict(d,subfeat,len,new_c,new_c_name):
    #new_c : new coloumn
    d_out = d
    for i in range(0,len):
        d_out[subfeat][i][new_c_name] = new_c[i,:,:]    
    return d_out
```

```python
def mergeDatasets(domain,source):
    org_path = "saved_features/test_feat_kinetics_"+domain+"_"+source+".pkl"
    audio_path = "audio_feat/"+domain+"-"+domain+"_"+source+"_Spec_TSN_.pkl"
    data = pd.DataFrame(pd.read_pickle(org_path))
    dic_org = pd.read_pickle(org_path)

    with open(audio_path, 'rb') as f:
        prova= hickle.load(f)

    new_feats = prova["\'features\'"]["\'Spec\'"][0]
    
    #Method 1
    #old_dataset = data.to_dict()
    old_dataset = dic_org
    new_dataset = addElementToDict(old_dataset,"features",data.shape[0],new_feats,"features_Audio")
    
    filename = "RGB_Audio_features/test_feat_rgb_audio_"+domain+"_"+source+".pkl"
    writePickle(filename,new_dataset)
    #df.to_pickle(filename)  
    
```

```python
#original features path:
domain = ["D1","D2","D3"]
source = ["train","test"]

for d in domain:
    for s in source:
        mergeDatasets(d,s)
```
